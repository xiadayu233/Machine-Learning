## Linear Regression
#### Assumptions:
1. linear relationship
2. Give $X$, $\epsilon_i$ are iid distributed with mean 0, (to make sure the estimate is unbiased)
3. Give $X$, $\epsilon_i$ are iid distributed with Variance $\sigma^2$, (to have the variance term for $\beta$)

#### explicit solution
1. loss function $E = \frac{1}{n}\sum_{i=1}^{n}(y_i -\alpha - x_i\beta)^2$
2. $\hat{\beta} = (X^TX)^{-1}X^TY$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (6.0, 4.0)
X = 2 * np.random.rand(100,1)
Y = 4 +3 * X+np.random.randn(100,1)

plt.scatter(X, Y)
plt.show()
```

```python
#build the model
m = 0 #beta
c = 0 #alpha
L = 0.001 # learning rate
epochs = 1000 #the number of iterations to perform gradient descent algo
n = len(X) #sample size
## perform gradient descent
for i in range(epochs):
    Y_pred = m*X+c
    D_m = (-2/n) * np.sum((Y-Y_pred)*X)     # derivative w.r.t. m
    D_c = (-2/n) * np.sum((Y-Y_pred))       # derivative w.r.t. c
    
    m = m - L*D_m                           #update m
    c = c - L*D_c                           #update c
```

```python
# make the predictions
Y_pred = m*X + c

plt.scatter(X, Y) 
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.show()
```
